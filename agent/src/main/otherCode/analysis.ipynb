```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme()
```

## Means progression

```python
def sample(N, mat):
    random_indices = np.random.randint(0, mat.shape[0], size=N)
    return mat[random_indices]

mean_progressions = pd.read_csv('chromosome-means.csv', header=None)
random_samples = sample(5, mean_progressions.values)
```

```python
x = list(range(1, len(random_samples[0]) + 1))
for index, random_sample in enumerate(random_samples):
    plt.plot(x, random_sample)
plt.xlabel('Generations')
plt.ylabel('Average Fitness')
plt.title('Progession of fitness through generations')

plt.show()
```

## Best individuals means 

```python
best_means = [float(line.rstrip()) for line in open('best-means.txt').readlines()]
best_means_df = pd.DataFrame(data={'means': best_means})
```

```python
sns.displot(best_means_df, x='means', kind='kde', bw_adjust=1.25)
plt.title('Best fitnesses distribution')
plt.xlabel('Fitness')
```

```python
sns.distplot(x=best_means_df['means'], bins=5)
plt.title('Best fitnesses distribution')
plt.xlabel('Fitness')
```
